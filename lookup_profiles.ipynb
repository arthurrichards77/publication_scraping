# Look-up Profiles

Workflow part 1: given a list of names, find the responsitory URLs for both UWE and UoB.  Save to XLS for verification and paper lookup.

```python
import pandas as pd
```

```python
from scrape_papers import PublicationScraper
scraper = PublicationScraper()
```

## University of Bristol

Taking names from the UoB SEMT Robotics team plus a few extras from other schools.

```python
uob_names = ['Andrew Conn',
             'Hermes Gadelha',
             'Martin Garrad',
             'Sabine Hauert',
             'Helmut Hauser',
             'Edmund Hunt',
             'Nathan Lepora',
             'Paul O\'Dowd',
             'Hemma Philamore',
             'Efi Psomopoulou',
             'Arthur Richards',
             'Jonathan Rossiter',
             'Antonia Tzemanaki',
             'Ben Ward-Cherrier',
             'Dandan Zhang',
             'Tom S Richardson',
             'Shane Windsor',
             'Walterio Mayol']
df_uob = pd.DataFrame({"name": uob_names})
df_uob
```

```python
df_uob['uob_url'] = [scraper.lookup_uob(n) for n in df_uob['name']]
df_uob
```

## UWE

This list initially guessed by Arthur

```python
uwe_names = ['Manuel Giuliani',
             'Tony Pipe',
             'Paul Bremner',
             'Mark Hansen',
             'Chris Harper',
             'Matthew Studley',
             'Martin Pearson',
             'Chris Melhuish',
             'Melvyn Smith',
             'Lyndon Smith',
             'Charlie Yang',
             'Carlos Cifuentes',
             'Marcela Munera',]
df_uwe = pd.DataFrame({"name": uwe_names})
df_uwe
```

```python
df_uwe['uwe_url'] = [scraper.lookup_uwe(n) for n in df_uwe['name']]
df_uwe
```

## Merge

Having looked up at each institution, merge the table and get everyone's Google Scholar IDs.

```python
df_uwe['institution'] = 'UWE'
df_uob['institution'] = 'UOB'
df = pd.merge(df_uob,df_uwe,how='outer')
df
```

```python
df['scholar_id'] = [scraper.lookup_scholar(n) for n in df['name']]
df
```

```python
df['scholar_url'] = 'https://scholar.google.com/citations?hl=en&user='+df['scholar_id']
df
```

```python
df.to_excel('results/profiles.xlsx')
```
