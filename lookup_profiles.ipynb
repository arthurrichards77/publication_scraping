```python
import pandas as pd
```

```python
from scrape_papers import PublicationScraper
scraper = PublicationScraper()
```

```python
names = ['Arthur Richards', 'Manuel Giuliani', 'Jonathan Rossiter', 'Paul Bremner',
         'Chris Melhuish', 
         'Sabine Hauert', 'Nathan Lepora', 'Mark Hansen', 'Antonia Tzemanaki',
         'Tony Pipe', 'Helmut Hauser', 'Andrew Conn', 'Ben Ward Cherrier', 'Chenguang Yang',
         'Charlie Yang', 'Chris Harper', 'Carlos Cifuentes']
```

```python
df = pd.DataFrame({"name": names})
df
```

```python
df['uob_url'] = [scraper.lookup_uob(n) for n in df['name']]
df
```

```python
df['uwe_url'] = [scraper.lookup_uwe(n) for n in df['name']]
df
```

```python
df['scholar_id'] = [scraper.lookup_scholar(n) for n in df['name']]
df
```

```python
df['scholar_url'] = 'https://scholar.google.com/citations?hl=en&user='+df['scholar_id']
df
```

```python
df.to_excel('profiles.xlsx')
```
