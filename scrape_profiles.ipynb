Step 2 of workflow: scrape the profiles already looked up and save a local database for each individual

```python
import pandas as pd
```

```python
from scrape_papers import PublicationScraper
scraper = PublicationScraper()
```

```python
profile_df = pd.read_excel('results/profiles.xlsx')
profile_df
```

## Grab UoB data

```python
for ii in range(len(profile_df)):
    this_name = profile_df.at[ii,'name'] 
    this_url = profile_df.at[ii,'uob_url']
    if type(this_url)==str:
        this_df = scraper.scrape_uob_research_info(this_url+'/publications/')
        this_file = 'results/' + this_name.replace(' ','_') + '_uob.xlsx'
        print(this_file)
        this_df.to_excel(this_file)
```

## Grab UWE data

```python
for ii in range(len(profile_df)):
    this_name = profile_df.at[ii,'name'] 
    this_url = profile_df.at[ii,'uwe_url']
    if type(this_url)==str:
        this_df = scraper.scrape_uwe_repository(this_url+'/outputs')
        this_file = 'results/' + this_name.replace(' ','_') + '_uwe.xlsx'
        print(this_file)
        this_df.to_excel(this_file)
```

## Scholar

For anyone with neither a UWE nor a UoB profile, try Scholar

```python
for ii in range(len(profile_df)):
    this_name = profile_df.at[ii,'name']
    this_url = profile_df.at[ii,'scholar_url']
    if type(this_url)==str:
        # skip if we have UWE or UOB URLs
        if type(profile_df.at[ii,'uwe_url'])==str:
            continue
        if type(profile_df.at[ii,'uob_url'])==str:
            continue
        this_df = scraper.scrape_scholar(profile_df.at[ii,'scholar_id'])
        this_file = 'results/' + this_name.replace(' ','_') + '_scholar.xlsx'
        print(this_file)
        this_df.to_excel(this_file)
```
