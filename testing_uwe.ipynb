```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://uwe-repository.worktribe.com/person/434603/user/outputs?page=1'
resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'html.parser')
```

```python
paper_tags = soup.find_all('blockquote')
```

```python
paper_tags[0].a.get('href')
```

```python
paper_tags[0].strong.text
```

```python
paper_tags[0].small.text[1:5]
```

```python
paper_tags[0].find_all(target='_blank')[0].text
```

```python
soup.title.text.strip()
```

```python
data_dict = {'title': [p.strong.text for p in paper_tags],
             'uwe_year_str': [p.small.text[1:5] for p in paper_tags],
             'uwe_detail_link': [p.a.get('href') for p in paper_tags]}
df = pd.DataFrame(data=data_dict)
df['uwe_source_url'] = url
df['date'] = pd.to_datetime(df['uwe_year_str'], format='mixed')
df['author_name'] = soup.title.text.strip()
df.head()
```
